```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset (make sure the path is correct)
titanic_df = pd.read_csv('/content/train.csv')
```

```python
# Data Cleaning
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
titanic_df['Cabin'].fillna('Unknown', inplace=True)
titanic_df['Sex'] = titanic_df['Sex'].astype('category')
titanic_df['Embarked'] = titanic_df['Embarked'].astype('category')
titanic_df['Pclass'] = titanic_df['Pclass'].astype('category')
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Plot the distribution of the 'Survived' variable
sns.countplot(x='Survived', data=titanic_df, palette='Set2')
plt.title('Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of the 'Pclass' variable
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', data=titanic_df, palette='Set2')
plt.title('Distribution of Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of the 'Sex' variable
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', data=titanic_df, palette='Set2')
plt.title('Distribution of Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of the 'Age' variable
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], kde=True, bins=30, color='teal')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plot the distribution of the 'Fare' variable
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Fare'], kde=True, bins=30, color='purple')
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
